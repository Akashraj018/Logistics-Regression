1. What is Simple Linear Regression (SLR)? Explain its purpose.

Simple Linear Regression models the relationship between X and Y by fitting a straight line (called the regression line) through the data points.
  Y = B0+B1x+e
Y - Dependent (response) variable — what you are trying to predict or explain.
X -Independent (predictor) variable — the variable you use for prediction
B0 - Intercept — the value of Y when X=0
B1 - Slope — how much Y changes for a one-unit change in X.
e -Error term — the difference between the observed and predicted values.

Purpose of SLR
Prediction – To predict the value of the dependent variable Y for a given value of X
Understanding Relationships – To quantify and test the strength and direction of the linear relationship between two variables.
Estimation – To estimate the parameters (B0 and B1) that best describe the relationship between the two variables.
  

2. What are the key assumptions of Simple Linear Regression?

The key assumptions of Simple Linear Regression (SLR) ensure that the model gives valid, unbiased, and efficient estimates. There are five main assumptions, which relate to the relationship between the variables and the behavior of the error terms.

Linearity - The relationship between the independent variable X and the dependent variable Y is Linear
The change in Y is proportional to the change in X
Use scatterplots or residual plots — if the points show a clear curve, the relationship is not linear

Independence of Errors
The residuals (errors) are independent of each other
The error for one observation should not depend on the error for another
Use the Durbin–Watson test to detect autocorrelation

Homoscedasticity (Constant Variance of Errors)
The variance of the residuals is constant across all levels of X
The spread of residuals should be roughly the same for all predicted values
Plot residuals vs. fitted values — the spread should be roughly uniform (no funnel shape).

Normality of Errors
The residuals are normally distributed
This assumption is important for hypothesis testing and confidence intervals.
Use a Q–Q plot or a histogram of residuals.

No Perfect Multicollinearity
Since SLR has only one predictor, this is automatically satisfied
(In multiple regression, it means predictors should not be perfectly correlated.)


3. Write the mathematical equation for a simple linear regression model and explain each term.

Y = B0 + B1X + e

Y - Dependent variable.
X - Independent variable
B0 - Intercept (constant term)
B1 - Slope (regression coefficient)
e - Error term (residual)

Example -Suppose you model the relationship between hours studied (X) and exam score (Y) as:
Exam Score=50+5×(Hours Studied)+e
B0=50: A student who studies 0 hours is expected to score 50
B1==5: For each additional hour studied, the exam score increases by 5 points (on average).
e: Represents random influences (e.g., test anxiety, sleep, etc.) that affect the score but aren’t related to hours studied.

5. What is the method of least squares in linear regression?

The method of least squares is the standard technique used to estimate the parameters of a linear regression model — that is, to find the best-fitting line through a set of data points.

The Model is Y = B0 + B1X + E
The method of least squares determines the values of B0 (intercept) and B1 (slope) that minimize the sum of the squared differences between the observed values and the predicted values of Y

For each data point (Xi,Yi)
predicted value is: Y^i =B0 + B1Xi

residual (error) is ei = Yi-Y^i

method of least squares minimizes the total squared error:
        S=i=1∑n​(Yi​−Yi​^​)2=i=1∑n​(Yi​−β0​−β1​Xi​)2

The values of B0 and B1 that minimize S give the best Fitting regression Line

7. Name and briefly describe three common evaluation metrics for regression models.

Here are three common evaluation metrics used to assess the performance of regression models (like linear regression):

1. Mean Absolute Error (MAE)
MAE=n1​i=1∑n​∣Yi​−Yi​^​∣
Description:

Measures the average absolute difference between the actual values (Yi) and the predicted values (Y^i)
It tells you, on average, how far your predictions are from the true values.
Interpretation:

A lower MAE indicates better model performance.
It’s easy to interpret because it’s in the same units as the target variable.

2. Mean Squared Error (MSE)
MSE=n1​i=1∑n​(Yi​−Yi​^​)2
Description:

Measures the average of the squared errors between actual and predicted values.
Squaring the errors penalizes larger deviations more heavily than smaller ones.

Interpretation:
A smaller MSE means a better fit.
It is sensitive to outliers since large errors are squared.

3. Coefficient of Determination
R2=1−∑(Yi​−Yˉ)2∑(Yi​−Yi​^​)2​
Description:

Represents the proportion of variance in the dependent variable that can be explained by the independent variable(s).
R2 ranges from 0 to 1.
nterpretation:

R2=1: Perfect prediction (model explains all variability).
R2=0: Model explains none of the variability (no predictive power).




8.  What is the purpose of the R-squared metric in regression analysis?

Purpose of the R-squared Metric in Regression Analysis R-squared metric also known as the coefficient of determination, is a statistical measure that indicates how well a regression model explains the variability of the dependent variable (Y) using the independent variable(s)(X).

Purpose
1.Evaluates Goodness of Fit
Indicates how well the regression line fits the observed data.
A higher R2 means the model’s predictions are closer to actual values
2.Compares Models
Helps compare the explanatory power of different models on the same dataset.
A model with a higher R2 typically fits better
3.Quantifies Explained Variability
Shows how much of the dependent variable’s variation is captured by the independent variable(s).

```python
 #Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

import numpy as np
from sklearn.linear_model import LinearRegression


X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

model = LinearRegression()

model.fit(X, y)

slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (β₁): {slope}")
print(f"Intercept (β₀): {intercept}")
```
